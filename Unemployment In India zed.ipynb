```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df.columns = df.colums = df.columns.str.strip()
```

```python
#Dropping irrelevant column
df = df.drop(columns=["Frequency"])
df.head()
```

```python
#Renaming Columns
df.rename(columns={'Region':'States',' Date':'Date',' Estimated Unemployment Rate (%)':'Estimated Unemployment Rate (%)',
                   ' Estimated Employed':'Estimated Employed',' Estimated Labour Participation Rate (%)':'Estimated Labour Participation Rate (%)',
                   'Region.1':'Region'},inplace=True)
```

```python
sns.pairplot(df, hue='Region')
plt.show()
```

```python
#Visualisation of uneployment per region
plt.figure(figsize=(15, 6))
plt.xticks(rotation=65) 
sns.barplot(data=df, x='Region', y='Estimated Unemployment Rate (%)', palette = 'Set3', order=df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values().index)
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate by Region')
plt.show()
```

```python
#Visualisation of uneployment per State
plt.figure(figsize=(15, 6))
plt.xticks(rotation=65) 
sns.barplot(data=df, x='States', y='Estimated Unemployment Rate (%)', palette = 'Set2', order=df.groupby('States')['Estimated Unemployment Rate (%)'].mean().sort_values().index)
plt.xlabel('States')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate by States')
plt.show()
```

```python
import datetime as dt
import calendar
#mapping date and time
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Month'] = df['Date'].dt.month
df['Month_int'] = df['Month'].apply(lambda x: int(x))
df['Month_name'] = df['Month_int'].apply(lambda x: calendar.month_abbr[x])
df.drop(columns='Month', inplace=True) 
```

```python
df
```

```python
# Grouping the data by 'States' and Estimated Unemployment Rate (%)
Unemployment_rate = df.groupby('States')['Estimated Unemployment Rate (%)'].sum()
Unemployment_rate = Unemployment_rate.sort_values(ascending=True)
Unemployment_rate
```

```python
#Plotting pie chart to show Estimated Unemployment Rate by States
plt.figure(figsize=(30, 15))
Unemployment_rate.head(9).plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Estimated Unemployment Rate by States', fontsize=16)
plt.ylabel('')
plt.show()
```
